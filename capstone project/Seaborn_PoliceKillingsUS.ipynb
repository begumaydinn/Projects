**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
# with seaborn

sns.boxplot(x="gender", y="age", data=df, hue="manner_of_death");
```

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn

plt.figure(figsize=(10,6))
sns.swarmplot("gender", y="age", data=df, hue="manner_of_death");
```

## 2: Plot 12 most dangerous cities

```python
df["city"].value_counts()[:12]
```

```python
# with seaborn barplot

plt.figure(figsize=(10,6))
ax = sns.barplot(ci= None, data=df, y=df["city"].value_counts()[:12], 
                 x=df["city"].value_counts()[:12].index)
ax.set_title("Most dangerous cities", color="blue", fontsize=15)
plt.xticks(rotation=45);
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
df["state"].value_counts()[:20]
```

```python
# with seaborn barplot

plt.figure(figsize=(8,5))
ax = sns.barplot(ci= None, data=df, x=df["state"].value_counts()[:20].index, y=df["state"].value_counts()[:20])
ax.set_title("Most dangerous states", color="blue", fontsize=15)
plt.xticks(rotation=45);
```

```python
# with seaborn countplot
```
